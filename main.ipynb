```python
import gensim
import os
import collections
import smart_open
import random
import re
```

```python
# file reader
def read_corpus(fname, tokens_only=False):
    with smart_open.smart_open(fname, encoding="iso-8859-1") as f:
    # with smart_open.smart_open(fname) as f:
        for i, line in enumerate(f):
            utterance = re.split(r'\t+', line)[1]
            print( utterance)
            #yield gensim.utils.simple_preprocess(utterance)
            yield gensim.models.doc2vec.TaggedDocument(gensim.utils.simple_preprocess(utterance), [i])
```

```python
filne_name ="data/CM_500__tutoringsession_utterance_info_with_speakerInfo_LM.txt"
read_corpus(filne_name)
train_corpus = list(read_corpus(filne_name))
```
